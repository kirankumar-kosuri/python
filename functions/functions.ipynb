```python
# defining functions
def even_or_odd(num):
    if num%2==0:
        print(f'Number is Even')
    else:
        print('Number is Odd')

# calling function
even_or_odd(24)
```

```python
def add (a,b):
    return (a+b)
result = add (1,1)
print(result)
```

```python
#Default Arguments
def greet (name):
    print (f"Hello '{name}' welcome to Python")
greet ('Kiran')
```

```python
# Positiona Arguments
def print_numbers(*args):
    for number in args:
        print (number)
print_numbers(1,2,"python")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print("Positional argument: '{val}'")
    for key,value in kwargs.items():
        print(f"'{key}':'{value}'")
print_details(1,2,3,4,5,6,7,8,9,"docker",name="kiran",age="23",city="hyd")
```
