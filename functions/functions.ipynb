```python
# Positiona Arguments
def print_numbers(*args):
    for number in args:
        print (number)
print_numbers(1,2,"python")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print("Positional argument: '{val}'")
    for key,value in kwargs.items():
        print(f"'{key}':'{value}'")
print_details(1,2,3,4,5,6,7,8,9,"docker",name="kiran",age="23",city="hyd")
```

```python
# Temperature conversion 
def convent_temperature (temp,unit):
# this function converts temperature between Celsius to Fahrenheit 
    if unit == 'C':
        return temp * 9/5 +32 # C to F
    elif unit == 'F':
        return (temp-32) *5/9 # F to C
    return None
print(convent_temperature(25,'C'))
print(convent_temperature(75,'F'))

```

```python
# checking the password strenght
def is_password_strong(password):
#this function checks weather the password is strong or not
    if len(password) < 8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char in '!@#$%^&*()_+{}[]' for char in password):
        return False
    return True
# calling function
print(is_password_strong("kiran12"))
print(is_password_strong("Kiran@1234"))


    


```

```python
def calculate_total_cost(cart):
    total_cost=0
    for item in cart:
        total_cost+=item['price']*item['quantity']
        return total_cost
# explain usage cart data
cart = [
    {'name': 'Apple' ,'price': 10 ,'quantity': 5},
    {'name': 'Mango' ,'price': 5, 'quantity': 3}

]
total_cost=(calculate_total_cost(cart))
print(total_cost)
```

```python
# checking the string is palindrome
def is_palindrome(s):
    """this is a function that checks wheather the palindrome or not"""
    s=s.lower().replace(" ","")
    return s==s[::-1]
print(is_palindrome("A man a plan a canal  Panama"))
print(is_palindrome("Hello"))
print(is_palindrome("racecar"))
```

```python
def count_word_frequencies(file_path):
    word_count={}
    with open (file_path,'r') as file:
        for line in file:
            words=line.split()
            for word in words:
                word=word.lower().strip('.,!?:;"|')
                word_count[word]=word_count.get(word,0)+1
    return word_count
file_path='vmr.txt'
word_frequency=count_word_frequencies(file_path)
print(word_frequency)
```
