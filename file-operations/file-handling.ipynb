```python
# read whole file
with open('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
# read file line by line
with open('example.txt','r') as file:
    for line in file:
        print(line.strip())
```

```python
# Overwrite 
with open('example.txt','w') as file:
    file.write('Kiran can teach Docker very well \n')
    file.write('Kiran can teach Kubernetes well')
```

```python
# Append
with open('example.txt','a') as file:
    file.write('\nPython is very east to learn')
```

```python
# add more lines to the file
lines=['\nLinux','\nAnsible','\nTerraform']
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
# Binary file
data = b'This is binary file example'
with open('binary.txt','wb') as file:
    file.write(data)
```

```python
# read binary file
with open('binary.txt','rb') as file:
    content=file.read()
    print(data)

```

```python
# copying one file content to another file content
with open('example.txt','r') as source_file:
    content=source_file.read()

with open('destination_file','w') as destination_file:
    destination_file.write(content)
```

```python
# Counting nunmber of words,lines and letters in file using python file haandling 
def count_text_file(file_path):
    with open(file_path,'r') as file:
        lines = file.readlines()
        lines_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        characters_count = sum(len(line) for line in lines)
    return lines_count, word_count,characters_count
file_path = 'example.txt'
lines,words,characters = count_text_file(file_path)
print(f'lines: {lines},words {words}, characters: {characters}')


```
