```python
# printing current working directory (cwd)
import os
cwd = os.getcwd()
print (f"The current working directory is '{cwd}'")
```

```python
# creating directory
directory = 'kiran'
os.mkdir(directory)
print (f"directory '{directory}' created")
```

```python
# list file and directories
items=os.listdir()
print (items)
```

```python
# print full path 
directory_name = 'pavan'
file_name = 'pavan.txt'
full_path=os.path.join(os.getcwd(),directory_name,file_name)
print(full_path)
```

```python
# checking files are exists are not
path = 'kiran1'
if os.path.exists(path):
    print(f"path '{path}' exists")
else:
    print(f"path '{path}' does not exists")
```

```python
# checking a path are directory
path = 'example.txt'
if os.path.isfile(path):
    print (f"the path '{path} is a file")
elif os.path.isdir(path):
    print(f"the path '{path}' is a directory")
else:
    print (f"path '{path} is file nor directory'")
```
